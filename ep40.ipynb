```python
class FOPCParser:
    def __init__(self, variables):
        self.variables = variables

    def parse_expression(self, expression):
        tokens = expression.split()
        stack = []

        for token in tokens:
            if token in self.variables:
                stack.append(self.variables[token])
            elif token == 'and':
                if len(stack) < 2:
                    raise ValueError("Insufficient operands for 'and' operator")
                operand2 = stack.pop()
                operand1 = stack.pop()
                result = operand1 and operand2
                stack.append(result)
            elif token == 'or':
                if len(stack) < 2:
                    raise ValueError("Insufficient operands for 'or' operator")
                operand2 = stack.pop()
                operand1 = stack.pop()
                result = operand1 or operand2
                stack.append(result)
            elif token == 'not':
                if len(stack) < 1:
                    raise ValueError("Insufficient operands for 'not' operator")
                operand = stack.pop()
                result = not operand
                stack.append(result)
            else:
                raise ValueError(f"Invalid token: {token}")

        if len(stack) != 1:
            raise ValueError("Invalid expression")

        return stack[0]

# Given variables
variables = {'p': True, 'q': True, 'r': False}

# Given expressions
expressions = [
    "p and q",
    "p or r",
    "not p",
    "q and (r or p)"
]

# Create FOPC parser
parser = FOPCParser(variables)

# Parse and evaluate each expression
for expression in expressions:
    try:
        result = parser.parse_expression(expression)
        print(f"Expression: '{expression}' - Result: {result}")
    except ValueError as e:
        print(f"Error: {e}")
```
