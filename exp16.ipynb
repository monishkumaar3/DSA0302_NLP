```python
import spacy

# Load the SpaCy English model
nlp = spacy.load("en_core_web_sm")

# Sentences for Named Entity Recognition
sentences = [
    "Apple Inc. is headquartered in Cupertino, California, and its CEO, Tim Cook, often delivers keynote speeches.",
    "The Eiffel Tower in Paris, France, is a popular tourist attraction."
]

# Perform Named Entity Recognition using SpaCy
for sentence in sentences:
    doc = nlp(sentence)

    # Display named entities and their labels
    print(f"Sentence: '{sentence}'")
    print("Named Entities:")
    for ent in doc.ents:
        print(f"{ent.text} - {ent.label_}")
    print()
```
