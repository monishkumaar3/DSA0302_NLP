```python
import nltk
from nltk.corpus import wordnet

# Download WordNet data (if not already downloaded)
nltk.download('wordnet')
nltk.download('punkt')

# Input sentences
sentences = [
    "The cat sat on the mat.",
    "The dog barked loudly.",
    "She played the piano beautifully."
]

# Function to explore WordNet for word meanings, synonyms, and relationships
def explore_wordnet(sentences):
    for sentence in sentences:
        words = nltk.word_tokenize(sentence)
        for word in words:
            synsets = wordnet.synsets(word)
            if synsets:
                print(f"Word: {word}")
                for synset in synsets:
                    print(f"Meaning: {synset.definition()}")  # Get meaning
                    print(f"Synonyms: {', '.join(synset.lemma_names())}")  # Get synonyms
                    hypernyms = synset.hypernyms()  # Get hypernyms (parent terms)
                    if hypernyms:
                        hypernyms_list = [', '.join(hypernym.lemma_names()) for hypernym in hypernyms]
                        print(f"Hypernyms: {', '.join(hypernyms_list)}")
                    hyponyms = synset.hyponyms()  # Get hyponyms (child terms)
                    if hyponyms:
                        hyponyms_list = [', '.join(hyponym.lemma_names()) for hyponym in hyponyms]
                        print(f"Hyponyms: {', '.join(hyponyms_list)}")
                    print("")

# Explore WordNet for the provided sentences
explore_wordnet(sentences)
```
