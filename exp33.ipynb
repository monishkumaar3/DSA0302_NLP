```python
import nltk
from nltk import word_tokenize, pos_tag
from nltk.chunk import RegexpParser

# Sample conversation
conversation = [
    "Good morning! How's the weather today?",
    "I heard it's going to be sunny and warm.",
    "Could you please send me the report by 3 PM?",
    "Of course, I'll send it over before the deadline.",
    "Do you know where the nearest post office is?",
    "The post office is two blocks down the street."
]

# Define dialog act patterns (you can extend these patterns for more dialog act types)
dialog_act_patterns = {
    'GREETING': r'^(Hello|Hi|Good morning|Good afternoon|Good evening|Hey)[,!\'".]*$',
    'INFORMATION_REQUEST': r'^([Ww]here|What|When|Who|How|Could you)[\w\s,?!\'".]*\??$',
    'INFORMATION_RESPONSE': r'^.*\bis\b.*$',
    'REQUEST': r'^([Cc]ould|Can|Please).*\b(send|give|provide)\b.*\bme\b.*\b.*\b\w+\b.*\bby\b.*\b\d+\s?(AM|PM)?\??$',
    'AFFIRMATION': r'^([Yy]es|[Oo]kay|[Ss]ure|[Aa]bsolutely)\b.*$',
    'THANKING': r'^([Tt]hank\syou|[Tt]hanks)\b.*$'
}

# Function to identify dialog acts in a given utterance
def identify_dialog_act(utterance):
    for act, pattern in dialog_act_patterns.items():
        if nltk.regexp_tokenize(utterance, pattern):
            return act
    return 'OTHER'

# Function to process the conversation and identify dialog acts
def process_conversation(conversation):
    dialog_acts = []
    for utterance in conversation:
        act = identify_dialog_act(utterance)
        dialog_acts.append((utterance, act))
    return dialog_acts

# Process the conversation and identify dialog acts for each utterance
identified_dialog_acts = process_conversation(conversation)

# Display identified dialog acts for each utterance
for utterance, act in identified_dialog_acts:
    print(f"Utterance: '{utterance}' - Dialog Act: '{act}'")
```
