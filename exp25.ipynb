```python
import re

def perform_regex_matching(text, regex):
    matches = re.findall(regex, text)
    return matches

# Text and regex pattern
text = "Regular expressions are a powerful tool for pattern matching in text. They provide a flexible way to search and manipulate strings."
regex_pattern = r"powerful|flexible|manipulate"

# Perform regex matching in the text
matched_words = perform_regex_matching(text, regex_pattern)

# Display matched words
print("Matched words:")
for word in matched_words:
    print(word)
```
