```python
import re

def find_pattern_in_text(pattern, text):
    matches = re.findall(pattern, text)
    return matches

# Regular expression pattern
pattern = r'\b\w{3}\b'  # Matches words that are exactly 3 characters long

# Input sentence
input_sentence = "The quick brown fox jumps over the lazy dog. The cat is also agile."

# Find and extract patterns in the input text using the specified regular expression
found_patterns = find_pattern_in_text(pattern, input_sentence)

# Display the found patterns
print("Found patterns matching the regular expression:")
for match in found_patterns:
    print(match)
```
