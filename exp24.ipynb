```python
def parse_expression(expression):
    # Replace 'and' with 'and', 'or' with 'or', and 'not' with 'not' to handle logic
    expression = expression.replace(' and ', ' and ').replace(' or ', ' or ').replace('not ', 'not ')

    # Evaluate the expression
    try:
        result = eval(expression, {'x': True, 'y': False})
        print(f"Expression: {expression}")
        print(f"Result: {result}")
    except Exception as e:
        print(f"Error evaluating the expression: {e}")

# Expressions to parse
expressions = [
    "x and y",
    "x or (not y)",
    "x and (y or (not x))"
]

# Parse each expression
for expr in expressions:
    parse_expression(expr)
```
